```python
DEPRESSION_PATH = "data/depression_dataset_reddit_cleaned.csv"
TDT_SPLIT = "80/10/10"

depression_word = helpers.read_file(datapath=DEPRESSION_PATH, ngram=2, by_character=False)
depression_char = helpers.read_file(datapath=DEPRESSION_PATH, ngram=2, by_character=True)

print(depression_word[0])
```

```python
# split data into train, dev, and test
train, dev, test = helpers.split(data=depression_word, dist=TDT_SPLIT)

print(f"len(data) = {len(depression_word)}")
print(f"Train size: {len(train)}, Dev size: {len(dev)}, Test size: {len(test)}")
```

```python
print("Train Data:")
helpers.get_stats(train)

print("Dev Data:")
helpers.get_stats(dev)

print("Test Data:")
helpers.get_stats(test)
```

```python
# TF-IDF and classifiers
data = helpers.read_file(datapath=DEPRESSION_PATH, ngram=2, tokenize=False, by_character=False)
clean_text = []
is_depression = []
for row in data:
    clean_text.append(row[0])
    is_depression.append(row[1])
```

```python
from multinomial_nb import MultinomialNaiveBayes

train_data = [(row[0], row[1]) for row in train]
dev_data = [(row[0], row[1]) for row in dev]
test_data = [(row[0], row[1]) for row in test]

nb_model = MultinomialNaiveBayes(alpha=1.0)
nb_model.fit(train_data)

print("=== Multinomial Naive Bayes ===")
print(f"Train accuracy: {nb_model.score(train_data):.4f}")
print(f"Dev accuracy:   {nb_model.score(dev_data):.4f}")
print(f"Test accuracy:  {nb_model.score(test_data):.4f}")
```
